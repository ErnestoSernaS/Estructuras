```python
from random import randrange

def DisplayBoard(board):
    print("+-------" * 3,"+",sep="")
    for fila in range(3):
        print("|       " * 3,"|",sep="")
        for columna in range(3):
            print("|   " + str(board[fila][columna]) + "   ",end="")
        print("|")
        print("|       " * 3,"|",sep="")
        print("+-------" * 3,"+",sep="")

def EnterMove(board):
    fv = False
    while not fv:
        movimiento = input("Ingresa tu movimiento: ") 
        fv = len(movimiento) == 1 and movimiento >= '1' and movimiento <= '9' 
        if not fv:
            print("Movimiento incorrecto, repitelo!")
            continue
        movimiento = int(movimiento) - 1 
        fila = movimiento // 3 
        columna = movimiento % 3
        tlsg = board[fila][columna]
        fv = tlsg not in ['O','X'] 
        if not fv:
            print("Campo ya ocupado, repita su entrada!")
            continue
    board[fila][columna] = 'O' 

def MakeListOfFreeFields(board):
    fmvl = []
    for fila in range(3): 
        for columna in range(3): 
            if board[fila][columna] not in ['O','X']: 
                fmvl.append((fila,columna)) 
    return fmvl


def VictoryFor(board,ind):
    if ind == "X":
        quien = 'yo'
    elif ind == "O":
        quien = 'tu'
    else:
        quien = None
    pasa1 = pasa2 = True  
    for fc in range(3):
        if board[fc][0] == ind and board[fc][1] == ind and board[fc][2] == ind:
            return quien
        if board[0][fc] == ind and board[1][fc] == ind and board[2][fc] == ind: 
            return quien
        if board[fc][fc] != ind: 
            pasa1 = False
        if board[2 - fc][2 - fc] != ind:
            pasa2 = False
    if pasa1 or pasa2:
        return quien
    return None

def DrawMove(board):
    libre = MakeListOfFreeFields(board) 
    tics = len(libre)
    if tics > 0:
        ess = randrange(tics)
        fila, columna = libre[ess]
        board[fila][columna] = 'X'

board = [ [3 * j + i + 1 for i in range(3)] for j in range(3) ]
board[1][1] = 'X' 
libre = MakeListOfFreeFields(board)
miturno = True 
while len(libre):
    DisplayBoard(board)
    if miturno:
        EnterMove(board)
        victoria = VictoryFor(board,'O')
    else:
        DrawMove(board)
        victoria = VictoryFor(board,'X')
    if victoria != None:
        break
    miturno = not miturno
    libre = MakeListOfFreeFields(board)

DisplayBoard(board)
if victoria == 'tu':
    print("Ganaste!")
elif victoria == 'yo':
    print("Te gane")
else:
    print("Empate!")
```
